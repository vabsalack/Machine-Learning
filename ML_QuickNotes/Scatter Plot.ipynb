***Scatter Plot***

```python
import matplotlib.pyplot as plt
```

The x array represents the age of each car.

The y array represents the speed of each car.

```python
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]

y = [99,86,87,88,111,86,103,87,94,78,77,85,86]
```

```python
plt.scatter(x, y)
plt.show()
```
